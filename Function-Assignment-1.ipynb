```python
def Elegiblity():
    Gender=input("Your Gender:")
    Age=float(input("Your Age: "))
    if (Gender=="Male" and Age>=21):
        print("Eligible")
    elif(Gender=="Female" and Age>=18):
        print("Eligible")
    else:
        print("Not Eligible")
```
